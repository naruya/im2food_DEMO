```python
# json0 の中10このファイルまとめてタイトル

import sys
import os
import pickle

import MeCab
m2 = MeCab.Tagger('-Ochasen')
m2.parse('') # これ重要!（最初のtextがバグるの防止）

path_pkl = '../cookpad/data_json/json0/'

pkl_files = os.listdir(path_pkl)
pkl_files.sort()

title_error = []

n_t, n_f = 0, 0

setubi = []
keiyogo = []
sahen = []
hijiritu = []
hukusika = []
daimeisi = []
kazu = []
dousihijiritu = []
tokusyu = []
nai = []
setuteki = []

for pkl_file in pkl_files:
    
    T = {} # タイトルの辞書 1pkl1辞書
    with open(path_pkl + pkl_file, 'rb') as f:
        menus = pickle.load(f)
    print(pkl_file) # あるpklについて
    for i, menu in enumerate(menus): # あるメニューについて
        key = pkl_file[:20] + '-' + '{0:07d}'.format(i)
        
        t = [] # あるメニューに含まれる名詞のリスト
        title = menu['name']
        node = m2.parseToNode(title)
        while node:
            if node.feature.split(',')[0] == '名詞':   # ↓○     ↓○      ↓×      ↓×          ↓× ...
                if not node.feature.split(',')[1] in ['一般', '固有名詞', '接尾','形容動詞語幹', 'サ変接続', '非自立', '副詞可能', '代名詞', '数', '動詞非自立的', '特殊', 'ナイ形容詞語幹', '接続詞的']:
                    print("    ", node.feature)
                if node.feature.split(',')[1] in ['接尾']:
                    setubi.append(node.surface)
                if node.feature.split(',')[1] in ['形容動詞語幹']:
                    keiyogo.append(node.surface)
                if node.feature.split(',')[1] in ['サ変接続']:
                    sahen.append(node.surface)
                if node.feature.split(',')[1] in ['非自立']:
                    hijiritu.append(node.surface)
                if node.feature.split(',')[1] in ['副詞可能']:
                    hukusika.append(node.surface)
                if node.feature.split(',')[1] in ['代名詞']:
                    daimeisi.append(node.surface)
                if node.feature.split(',')[1] in ['数']:
                    kazu.append(node.surface)
                if node.feature.split(',')[1] in ['動詞非自立的']:
                    dousihijiritu.append(node.surface)
                if node.feature.split(',')[1] in ['特殊']:
                    tokusyu.append(node.surface)
                if node.feature.split(',')[1] in ['接続詞的']:
                    setuteki.append(node.surface)
                if node.feature.split(',')[1] in ['ナイ形容詞語幹']:
                    nai.append(node.surface)
                t.append(node.surface)
            node = node.next
        
        if len(t) == 0:
            title_error.append(title)
            n_f += 1
            continue
        
        T.update({key:t}) # 1メニューずつ突っ込む
        #print(t)
        #print()
        #print(key)
        n_t += 1
    #with open('recipe_title/' + "recipe_title-" + pkl_file[5:20] + '.pkl', 'wb') as f:
    #    pickle.dump(T, f)
    break
print("done")

n_f = len(title_error)

print("recipe_error")
n = min(len(title_error), 10)
for i in range(n):
    print(" - ", title_error[i])
print("採用: ", n_t , "レシピ, 不採用: ", n_f, "レシピ")

print('\n接尾\n', set(setubi))
print('\n形容動詞語尾\n', set(keiyogo))
print('\nサ変接続\n', set(sahen))
print('\n非自立\n', set(hijiritu))
print('\n副詞可能\n', set(hukusika))
print('\n代名詞\n', set(daimeisi))
print('\n数\n', set(kazu))
print('\n動詞非自立的\n', set(dousihijiritu))
print('\n特殊\n', set(tokusyu))
print('\n接続詞的\n', set(tokusyu))
print('\nナイ形容詞語幹\n', set(tokusyu))
```

# タイトルを綺麗にしてみる

```python
import sys
import os
import pickle

import re

import MeCab
m2 = MeCab.Tagger('-Ochasen')
m2.parse('') # これ重要!（最初のtextがバグるの防止）

path_pkl = '../cookpad/data_json/json0/'

pkl_files = os.listdir(path_pkl)
pkl_files.sort()

title_error = []

n_t, n_f = 0, 0

def parse(text):
    TOKENS = re.compile('[a-zA-Zａ-ｚＡ-Ｚ0-9０-９]')
    results = TOKENS.findall(text)
    return results

T = {} # タイトルの辞書 1pkl1辞書
for pkl_file in pkl_files:
    
    
    with open(path_pkl + pkl_file, 'rb') as f:
        menus = pickle.load(f)
    print(pkl_file) # あるpklについて
    for i, menu in enumerate(menus): # あるメニューについて
        key = pkl_file[:20] + '-' + '{0:07d}'.format(i)
        
        t = [] # あるメニューに含まれる名詞のリスト
        title = menu['name']
        print(title)
        node = m2.parseToNode(title)
        while node:
            if node.feature.split(',')[0] == '名詞':   # ↓○     ↓○      ↓×      ↓×          ↓× ...
                if node.feature.split(',')[1] in ['一般', '固有名詞']:#, '接尾','形容動詞語幹', 'サ変接続', '非自立', '副詞可能', '代名詞', '数', '動詞非自立的', '特殊', 'ナイ形容詞語幹', '接続詞的']:
                    #print("    ", node.feature)
                    a = node.surface
                    
                    trash = []
                    for e in parse(a):
                        trash.append(e)
                    
                    
                    if len(trash) == 0: # 数字やローマ字が含まれてなかったらその単語はおｋそれ以外はダメ
                        t.append(a)
                        
            node = node.next
        
        if len(t) == 0:
            title_error.append(title)
            n_f += 1
            continue
        
        T.update({key:t}) # 1メニューずつ突っ込む
        print("    =>", t, key)
        n_t += 1
        if n_t >= 10:
            break
    

print("done")

print(len(T))

n_f = len(title_error)

print("recipe_error")
n = min(len(title_error), 10)
for i in range(n):
    print(" - ", title_error[i])
print("採用: ", n_t , "レシピ, 不採用: ", n_f, "レシピ")
```
